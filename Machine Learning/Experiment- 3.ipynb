```python
import pandas as pd
import matplotlib.pyplot as plt
#Scikit Learn Library
from sklearn import datasets
#print(datasets)
#load the Iris dataset in dictionary values which will be in the key value pair
iris=datasets.load_iris()
#print(iris)
#print(iris.keys())
#print(iris["feature_names"])
#print(iris["data"])
#print(iris['DESCR'])
#place the data in the table format(DataFrame)
df=pd.DataFrame(iris['data'],columns=iris['feature_names'])
df
```

```python
#rescale all the value in the same unit,(formula is xi-mean/SD) mean is always set to zero & Standard DEviation to 1
from sklearn.preprocessing import StandardScaler
#this helps to rescale to the same unit
scalar=StandardScaler()
scalar.fit(df)
#transforms to the required scale
scaled_data=scalar.transform(df)
print(scaled_data)
```

```python
#apply pca by importing
from sklearn.decomposition import PCA
#specify how many components we want to keep when creating the PCA object
pca=PCA(n_components=2)
#pca=PCA(n_components=3)
#pca=PCA(n_components=4)
pca.fit(scaled_data)
#when variance is higher, the points are not overlapped but clearly projected on pca
print("variance value:",pca.explained_variance_ratio_)
x_pca=pca.transform(scaled_data)
#x_pca.shape
print(x_pca)
```

```python
#now plot the figure
plt.figure(figsize=(12,10))
y=iris.target
target_names=iris.target_names
print(y)
#plt.scatter(x_pca[:,0],x_pca[:,1],c=iris['target'],cmap='plasma')
for i,target_name in enumerate(target_names):
    plt.scatter(x_pca[y==i,0],x_pca[y==i,1],label=target_name)
plt.title("PCA on Iris Dataset")
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.legend()
plt.show()
```
