```python
import numpy as np
import pandas as pd

#ignore all warnings

import warnings
warnings.filterwarnings("ignore")

#Step 1: Generate the 100 random values as dataset

values=np.random.rand(100)#keep generates values between 0-1
#values=np.random.seed(100)#generates same values everytime we run
labels=[]

for x in values[:50]:
    if x<=0.5:
        labels.append('Class1')#add class2 next to the existing values
    else:
        labels.append('Class2')
#print(values)
#print(labels)
```

```python
#Adding the none values for the extra 50 values as None in tables
labels+=[None]*50
#print(labels)
data={'Point': [f"x{i+1}" for i in range(100)],"Values": values,"Labels":labels}
#print(data)
df=pd.DataFrame(data)
#df
```

```python
#split data into labeled and unlabeled
labeled_df=df[df["Labels"].notna()]#notna function returns not null rows
X_train=labeled_df[["Values"]]
Y_train=labeled_df[["Labels"]]

unlabeled_df=df[df["Labels"].isna()]#isna function returns null rows
X_test=unlabeled_df[["Values"]]

#generate true values for testing data(for accuracy calculations)
true_labels=["Class1" if x<=0.5 else "Class2" for x in values[50:]]
#true_labels
```

```python
#(Create knn-k nearest classifier)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
k_values=[1,2,3,4,5,20,30]
#classify using k nearest neighbors
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)#train the models
    prediction=knn.predict(X_test)#test
    print(f"Accuracy for k={k}")
    print(classification_report(true_labels,prediction))
```
