```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_boston
import warnings
warnings.filterwarnings("ignore")

#dat=fetch_california_housing()
#dat=fetch_california_housing(as_frames=True)
dat=load_boston()
#print(dat)
Y=dat.target
#print(y)
df=pd.DataFrame(dat.data,columns=dat.feature_names)
df

```

```python
X=df[['INDUS']]
#split the data
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(X_train,Y_train)#Completed Model training
y_pred=lg.predict(X_test)
#lg.predict(X_test)
```

```python
#plotting the results
plt.scatter(X_test,Y_test,color="blue")
plt.plot(X_test['INDUS'],y_pred,color='red')
plt.title('Linear Regression')
plt.show()
```
